Цель работы: на данной практической работе необходимо использовать датасет с прошлой практической работы и визуализировать данные в виде графиков. Необходимо сделать минимум 3 линейных графика, минимум 3 столбчатых диаграммы, минимум 2 круговых диаграммы, минимум 2 диаграммы рассеяния, минимум 3 гистограммы. 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 

anime = pd.read_csv("anime.csv")

plt.plot(anime.columns.values.tolist (), anime.notnull().sum(), marker='o', markersize=3, label='Количество непустых значений')
plt.xlabel('Названия колонок')
plt.ylabel('Заполненные ячейки')
plt.grid()
plt.legend()
plt.show()
```

Создаём обычный линейный график с соотношением колонок к непустым значениям. После построения можно увидеть, что у большинства колонок практически все значения непустые кроме рейтинга, там довольно много незаполненных ячеек. 

```python
x = anime["type"].unique().tolist()
y = []

for item in x:
    y.append(len(anime[anime['type']==item]))

plt.plot(x, y, marker='o', markersize=3, color="pink", label='Количество типов аниме')
plt.xlabel('Названия типов')
plt.ylabel('Количество')
plt.grid()
plt.legend()
plt.show()
```

Создаём обычный линейный график с соотношением типов аниме к их количеству в датасете. После построения можно увидеть, что больше всего у нас TV, OVA на 2 месте, Movie на 3 месте, ONA и Music +- одинаково, NaN полностью отсутствует.

```python
plt.plot(list(range(1, 51)), anime['rating'].head(50), marker='o', markersize=3, color="red", label='Рейтинг аниме')
plt.xlabel('Номер аниме в датасете')
plt.ylabel('Рейтинг')
plt.grid()
plt.legend()
plt.show()
```

Создаём обычный линейный график с соотношением номера аниме в датасете к его рейтингу. После построения можно увидеть, что есть аниме у которых нет рейтинга, а у остальных значения +- совпадают.

```python
a = [1, 3, 5, 7, 9 ,10]
x = []
y = []
i = 0
while i<5:
    y.append(len(anime[(anime['rating']>=a[i]) &(anime['rating']<=a[i+1])]))
    x.append(f"{a[i]} - {a[i+1]}")
    i +=1

plt.bar(x, y, color="purple", label='Рейтинг аниме')
plt.xlabel('Рейтинг')
plt.ylabel('Количество аниме')
plt.legend()
plt.show()
```

Создаём столбчатую диаграмму с сотношением промежутков рейтинга к количеству анмешек с данным рейтингом. После построения можно увидеть, что больше всего у нас анимешек с рейтингом 5-7, на втором месте - 7-9, на третем - 3-5 и рейтинга 1-3 и 9-10 практически нет.

```python
i = 0
x = []
y = []
anime["rating"] = np.where(anime["rating"].isnull(), 0, anime["rating"])

for item in anime['rating'].tail(50):
    y.append(item)
    x.append(i)
    i += 1

plt.bar(x, y, color="yellow", label='Рейтинг аниме')
plt.plot(x, y, marker='o', markersize=3, color="orange", label='Количество типов аниме')
plt.xlabel('Номер аниме в датасете')
plt.ylabel('Рейтинг')
plt.legend()
plt.show()
```

Создаём столбчатую диаграмму с соотношением послених анимешек в датасете к их рейтингу. После можно увидеть, что диаграмма в основном колеблится в рамках от 3 до 6 баллов, но есть и исключения. 

```python
anime = pd.read_csv("anime.csv")

plt.bar(list(range(1, 51)), anime['members'].head(50), color="pink")
plt.plot(list(range(1, 51)), anime['members'].head(50), marker='o', markersize=3, color="red")
plt.xlabel('Номер аниме в датасете')
plt.ylabel('Количество просмотров в млн.')
plt.show()
```

Создаём столбчатую диаграмму с сотношением первых 50 аниме к количеству просмотров. После можно увидеть значения очень сильно разнятся. Максимальное значение чуть больше миллиона, а самое маленькое несколько тысяч просмотров. 

```python
labels = anime["type"].unique().tolist()
vals = []

for item in labels:
    vals.append(len(anime[anime['type']==item]))

plt.pie(vals, labels=labels, autopct='%1.1f%%')
plt.title("Соотношение типов аниме")
plt.show()
```

Создаём круговую диаграмму с сотношением типов аниме в датасете. Из получившейся диаграммы можно увидеть, что TV и OVA +- равны и их больше всего, Special на втором месте, ONA и Music тоже +- равны.

```python
a = [1, 3, 5, 7, 9 ,10]
labels = []
vals = []
i = 0
while i<5:
    vals.append(len(anime[(anime['rating']>=a[i]) &(anime['rating']<=a[i+1])]))
    labels.append(f"{a[i]} - {a[i+1]}")
    i +=1
plt.pie(vals, labels=labels, autopct='%1.1f%%')
plt.title("Соотношение аниме с определенным рейтиногом")
plt.show()
```

Создаём круговую диаграмму с сотношением промежутков рейтинга к количеству анмешек с данным рейтингом. После построения можно увидеть, что больше всего у нас анимешек с рейтингом 5-7, на втором месте - 7-9, на третем - 3-5 и рейтинга 1-3 и 9-10 практически нет.

```python
plt.figure(figsize=(10, 6))
plt.scatter(list(range(1, 61)), anime['rating'].head(60), color="green")
plt.xlabel('Номер аниме')
plt.ylabel('Рейтинг')
plt.title("Рейтинг у первых 60 аниме")
plt.show()
```

Создаём график рассеивания с соотношением первых 60 анимешек в датасете к их рейтингу. После можно увидеть, что график у нас довольно плавно убывает, но разрыв значений довольно маленький. 

```python
plt.figure(figsize=(10, 6))
plt.scatter(list(range(1, 61)), anime['members'].head(60), color="blue")
plt.xlabel('Номер аниме')
plt.ylabel('Просмотры в млн.')
plt.title("Просмотры у первых 60 аниме")
plt.show()
```

Создаём график рассеивания с сотношением первых 60 аниме к количеству просмотров. После можно увидеть значения очень сильно разнятся. Максимальное значение чуть больше миллиона, а самое маленькое несколько тысяч просмотров. 

```python
plt.figure(figsize=(10, 6))
plt.hist(anime['rating'], bins=30, color=["green"], edgecolor='yellow', alpha=0.7, label='Распределение количества просмотров аниме')
plt.legend()
plt.show()
```

Создаём гистограмму в которой будем сравнивать рейтинг всех аниме в датасете. После построения можно увидеть, что больше всего у нас аниме с рейтингом в диапозоне 6-7, а меньше всего аниме с рейтингом 0-2 и 9-10.

```python
plt.figure(figsize=(10, 6))
plt.hist(anime['members'].tail(60), bins=30, color=["pink"], edgecolor='orange', alpha=0.7, label='Распределение просмотров последних 60 аниме')
plt.legend()
plt.show()
```

Создаём гистограмму в которой будем сравнивать простмотры последних 60 аниме в датасете. После построения можно увидеть, что больше всего аниме с минимальным количеством просмотров и чем больше увеличивается количество просмотров, тем меньше там аниме.

```python
plt.figure(figsize=(10, 6))
plt.hist(anime['members'].head(60), bins=30, color=['aquamarine'], edgecolor='cyan', alpha=0.7, label='Распределение просмотров первых 60 аниме')
plt.legend()
plt.show()
```

Создаём гистограмму в которой будем сравнивать просмотры первых 60 аниме в датасете. После построения можно увидеть, что больше всего аниме с минимальным количеством просмотров и чем больше увеличивается количество просмотров, тем меньше там аниме.
